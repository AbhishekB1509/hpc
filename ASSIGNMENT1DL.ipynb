```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,mean_absolute_error
from sklearn.preprocessing import StandardScaler
import seaborn as sns
```

```python
correlation_metrics=df.corr()
plt.figure(figsize=(12,8))
sns.heatmap(correlation_metrics,annot=True)
plt.title("correlation metrics")
plt.show()
```

```python
x=df.drop(['medv'],axis=1)
y=df['medv']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
scale=StandardScaler()
x_train=scale.fit_transform(x_train)
x_test=scale.transform(x_test)
```

```python
linear_regression=LinearRegression()
linear_regression.fit(x_train,y_train)
```

```python
y_pred=linear_regression.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model=Sequential()
```

```python
model.add(Dense(64,input_dim=x_train.shape[1],activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(1))
model.compile(optimizer='adam',loss='mean_squared_error',metrics=['mae'])
history = model.fit(x_train, y_train, epochs=50, validation_split=0.05)
```

```python
output = model.evaluate(x_test,y_test)
output
```

```python
print(f"Mean Squared Error: {output[0]}",f"Mean Absolute Error: {output[1]}")
```
